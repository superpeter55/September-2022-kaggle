# Preprocessing

```python
# Setting up X and Y for train and Test 
X_train = train[train.columns.drop(['row_id','num_sold'])]
y_train = train['num_sold']
X_test = test[test.columns.drop(['row_id'])]
```

```python
### Covid feature seems to hurt results
# Fuction that adds 2020 feature
def after_2020(date):
    year, month, day = date.split('-')
    if int(year) >= 2020:
        return 1
    return 0
```

```python
# Between Christmas and New Years
def christmas_new_year(date):
    year, month, day = date.split('-')
    if int(month) == 12 and int(day) > 25:
        return 1
    if int(month) == 1 and int(day) == 1:
        return 1
    return 0
```

```python
def period_location(date, product):
    """Returns sin and cos period location given date and product"""
    # Determining Period Length
    if product == "Kaggle for Kids: One Smart Goose":
        period_frac = two_year_period(date)
    else:
        period_frac = one_year_period(date)
    # Calculating sin and cos location
    sin_location = np.sin(2*np.pi*period_frac)
    cos_location = np.cos(2*np.pi*period_frac)
    
    return sin_location, cos_location
    
    
def one_year_period(date):
    """Returns ratio of period that is completed for one-year periods"""
    period_len = 365
    return day_of_current_year(date) / period_len
    
def two_year_period(date):
    """Returns ratio of period that is completed for two-year periods"""
    period_len = 730
    year, month, day = date.split('-')
    if year == "2018" or year == "2020":
        return (365 + day_of_current_year(date)) / period_len
    else:
        return day_of_current_year(date) / period_len
    
def day_of_current_year(date):
    year, month, day = date.split('-')
    year = int(year)
    day = int(day)
    month = int(month)
    # Dict with days in months
    days_in_months = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30,
                     7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
    
    # Counter for completed days in year
    completed_month_days = 0
    for key in days_in_months:
        if key < month:
            completed_month_days += days_in_months[key]
    # Leap year
    if year == 2020 and month > 2:
        completed_month_days += 1
    
    return completed_month_days + day
```

#### Adding Day of Week

0 = monday, 1 = tuesday, etc

```python
def year_month_day(df):
    """Adds year month day and day of week
    Drops date column"""
    df = df.copy()
    df['year'] = df['date'].map(lambda x: x.split('-')[0])
    df['month'] = df['date'].map(lambda x: int(x.split('-')[1]))
    df['day_num'] = df['date'].map(lambda x: x.split('-')[2])
    df['day_name'] = df['date'].map(lambda x: datetime.datetime(int(x.split('-')[0]), 
                                                                int(x.split('-')[1]),
                                                                int(x.split('-')[2])).weekday())
    df['day_of_week_sin'] = np.sin(df['day_name'] * (2 * np.pi / 7))
    df['day_of_week_cos'] = np.cos(df['day_name'] * (2 * np.pi / 7))
    df['day_of_month_sin'] = np.sin(df['month'] * (2 * np.pi / 12))
    df['day_of_month_cos'] = np.cos(df['month'] * (2 * np.pi / 12))
    df['cycle_location_sin'] = df.apply(lambda row: period_location(row['date'],row['product'])[0], axis = 1)
    df['cycle_location_cos'] = df.apply(lambda row: period_location(row['date'],row['product'])[1], axis = 1)
    df['after_2020'] = df.apply(lambda row: after_2020(row['date']), axis = 1)
    df['christmas_time'] = df.apply(lambda row: christmas_new_year(row['date']), axis = 1)
    df = df[df.columns.drop(['date', 'day_name', 'month','day_num'])]
    
    return df
```

```python
X_train = year_month_day(X_train)
X_test = year_month_day(X_test)
```

```python
# Splitting Categorical and Numerics
categoricals = ['country', 'store', 'product']
numerics = ['year', 'day_of_week_sin', 'day_of_week_cos', 'day_of_month_sin', 'day_of_month_cos',
            'after_2020','christmas_time', 'cycle_location_sin', 'cycle_location_cos']

X_train_cat = X_train[categoricals]
X_train_num = X_train[numerics]

X_test_cat = X_test[categoricals]
X_test_num = X_test[numerics]
```

```python
# One hot encoding and combining
enc = OneHotEncoder()

X_train_cat = enc.fit_transform(X_train_cat)
X_test_cat = enc.fit_transform(X_test_cat)
```

```python
X_train = pd.concat([pd.DataFrame(X_train_cat.todense()), X_train_num], axis = 1)
X_test = pd.concat([pd.DataFrame(X_test_cat.todense()), X_test_num], axis = 1)
```

# Model

```python
# Defining Classifier
# Memory errors cause by max depth issues
clf = RandomForestClassifier(max_depth = 10)
```

```python
# Fitting and Predicting
# Memory Errors due to max depth issues
clf.fit(X_train, y_train)
```

```python
num_sold = clf.predict(X_test)
```

# Submission File

```python
def write_submission(row_id, num_sold):
    """function to write submission from predictions"""
    ans = pd.concat([row_id, num_sold], axis = 1)
    ans.to_csv('submission.csv', index = False)
```

```python
write_submission(row_id, pd.Series(num_sold, name = 'num_sold'))
```
